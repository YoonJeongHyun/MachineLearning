```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

import pandas as pd
```

```python
breast_df = pd.read_csv("data/breast_cancer.csv")
breast_df.head()
```

```python
breast_df["target"].value_counts().index
```

```python
encoder1 = LabelEncoder()
# 데이터를 학습시킨다.
encoder1.fit(breast_df["target"])
# 학습한 것을 토대로 변환한다.
df_enc1 = encoder1.transform(breast_df["target"])

df_enc1
```

```python
# 복원
df_enc2 = encoder1.inverse_transform(df_enc1)
```

```python
df_enc2
```

```python
# 표준화 작업을 위해 결과 칼럼은 제외
X = breast_df.drop("target", axis=1)
X
```

```python
# 표준화
scaler1 = StandardScaler()
# 학습한다.
scaler1.fit(X)
# 변환한다.
scaler_df = scaler1.transform(X)

scaler_df
```
